## Extraction function per file type

```python
def extract_pdf(file_path):
    doc = fitz.open(file_path)
    text = ""
    for page in doc:
        text += page.get_text("text") + "\n"
    return text


# testing pdf extraction
pdf_text = extract_pdf("..\data\pdfs\Aurélien Géron - Hands-On Machine Learning with Scikit-Learn, Keras, and TensorFlow_ Concepts, Tools, and Techniques to Build Intelligent System (2022, O'Reilly Media) - libgen.li.pdf")

print(pdf_text[:1000]) #print first 1000 chars
len(pdf_text)
```

```python
def extract_epub(file_path):
    book = epub.read_epub(file_path)
    text = []
    for item in book.get_items():
        if item.get_type() == ebooklib.ITEM_DOCUMENT:
            soup = BeautifulSoup(item.get_content(), 'html.parser')
            
            # 1. Use strip=True to remove leading/trailing whitespace from the whole item
            # 2. Use a separator (like \n) so words don't get smashed together
            item_text = soup.get_text(separator="\n", strip=True)
            
            # Only add to our list if there is actually content
            if item_text:
                text.append(item_text)
                
    # Join with a single newline to keep things tight
    return "\n".join(text)
# Try it now!

# Try it!
epub_text = extract_epub("../data/epubs/Fluent Python, 2nd Edition - Luciano Ramalho - 2.epub")
print(epub_text[:1000])

len(epub_text)
```

```python
def extract_mobi(file_path):
    # Extracts to a temp folder and gives us the path
    tempdir, filepath = mobi.extract(file_path)
    with open(filepath, 'r', encoding='utf-8', errors='ignore') as f:
        content = f.read()
    
    # Use BeautifulSoup with a specific separator to avoid "smashing" words
    soup = BeautifulSoup(content, 'html.parser')
    
    # 1. strip=True removes leading/trailing spaces
    # 2. separator="\n" preserves paragraph breaks
    text = soup.get_text(separator="\n", strip=True)

    # MOBI extraction often leaves triple or quadruple newlines. 
    # Let's clean that up so we don't waste context space.

    lines = [line.strip() for line in text.split('\n') if line.strip()]
    return "\n".join(lines)

mobi_text = extract_mobi("../data/mobis/François Chollet - Deep Learning with Python (2021, MANNING Publications) - libgen.li.mobi")
print(mobi_text[:1000])
len(mobi_text)
```

## EXTRACTION FUCTION FOR ALL THE BOOKS

```python
# fucntion to extract all the books form the data folder and store then book-by book

def ingest_all_books(base_data_path="../data"):
    processed_docs = []

    # Mapping folders to their respective extractors (from our previous cells)
    # Note: Make sure those functions are defined in your notebook!

    format_map = {
        "pdfs": extract_pdf,
        "epubs": extract_epub,
        "mobis": extract_mobi
    }    

    for folder, extractor_func in format_map.items():
        folder_path = os.path.join(base_data_path, folder)

        # Check if the folder exists

        if not os.path.exists(folder_path):
            print(f"Skipping {folder}: FOlder not found.")
            continue

        # Get all files in this folder (ignoring .gitkeep and hidden files)
        files = [f for f in os.listdir(folder_path) if not f.startswith(".")]

        print(f"Processing {len(files)} files in {folder}...")

        for filename in files:
            file_path = os.path.join(folder_path, filename)

            try:
                # Run the extractor
                text = extractor_func(file_path)

                # Create the Document Object with Metadata

                doc_entry = {
                    "content": text,
                    "metadata": {
                        "filename": filename,
                        "extension": os.path.splitext(filename)[1],
                        "source_folder": folder,
                        "char_count": len(text)
                    }
                }
                processed_docs.append(doc_entry)
                print(f"  ✓ Successfully extracted: {filename} ({len(text)} characters)")

            except Exception as e:
                print(f" ✗ Error extracting {filename}: {str(e)}")

    return processed_docs


# --- 2. Execute and Inspect ---
# Run the ingestion

all_books = ingest_all_books()

# Quick summary

print(f"\n--- Total books processed: {len(all_books)}---")
for doc in all_books:
    meta = doc['metadata']
    print(f"[{meta['extension'].upper()}] {meta['filename']} - {meta['char_count']} chars")

```

```python
import re
import unicodedata
import os

# --- 1. THE UNIVERSAL WORKER ---
def clean_text_pipeline_master(text, metadata=None):
    """A production-grade cleaner that dynamically handles any book."""
    
    # --- PHASE A: Character Normalization ---
    # NFKC merges similar characters and fixes weird PDF encoding
    text = unicodedata.normalize('NFKC', text)
    # Smart quotes/dashes normalization (Vital for search accuracy)
    text = (text.replace('“', '"').replace('”', '"')
                .replace('‘', "'").replace('’', "'")
                .replace('—', '-').replace('–', '-'))

    # --- PHASE B: Strategy A (Dynamic Header/Footer Removal) ---
    if metadata and 'filename' in metadata:
        # 1. Extract a clean version of the title from the filename
        # e.g., "Python_Books_v1.pdf" -> "Python Books v1"
        base_name = os.path.splitext(metadata['filename'])[0]
        dynamic_title = base_name.replace('_', ' ').replace('-', ' ')
        
        # 2. Case-insensitive removal of that title if it sits on its own line
        # (This kills headers that repeat on every single page)
        text = re.sub(rf"^\s*{re.escape(dynamic_title)}\s*$", "", text, flags=re.MULTILINE | re.IGNORECASE)

    # --- PHASE C: Static Junk Removal (The Catch-Alls) ---
    static_junk = [
        r"Copyright © \d{4}.*", 
        r"All rights reserved",
        r"ISBN-13:.*",
        r"NO STARCH PRESS" # Add other common publishers here if needed
    ]
    for junk in static_junk:
        text = re.sub(junk, "", text, flags=re.IGNORECASE)

    # --- PHASE D: Structural Cleanup ---
    # Table of Contents dots
    text = re.sub(r"\.\s?\.(\s?\.)+", "", text)
    # Standalone numbers (Page numbers)
    text = re.sub(r"^\s*([0-9ivx]+)\s*$", "", text, flags=re.MULTILINE)
    # PDF Hyphenation (jo-ining words broken by newlines)
    text = re.sub(r"(\w+)-\n(\w+)", r"\1\2", text)
    # Bullet Point Normalization (Ensures search consistency)
    text = re.sub(r"^[ \t]*[\*\-\•][ \t]+", "• ", text, flags=re.MULTILINE)

    # --- PHASE E: Whitespace & Control Characters ---
    # Strip invisible control characters (like \x0c form-feeds)
    text = "".join(ch for ch in text if unicodedata.category(ch)[0] != "C")
    # Collapse multiple horizontal spaces (tabs/spaces)
    text = re.sub(r"[ \t]+", " ", text)
    # Collapse 3+ newlines into exactly 2 (keeps paragraphs, kills gaps)
    text = re.sub(r"\n{3,}", "\n\n", text)
        
    return text.strip()

# --- 2. THE BATCH MANAGER ---
def run_final_cleaning(books_list):
    print(f"🚀 Cleaning Library: {len(books_list)} books found.\n")
    
    for i, book in enumerate(books_list):
        raw_size = len(book['content'])
        
        # We pass the metadata to the worker for 'Strategy A'
        book['content'] = clean_text_pipeline_master(book['content'], book['metadata'])
        
        clean_size = len(book['content'])
        reduction = (1 - (clean_size / (raw_size or 1))) * 100
        
        print(f"[{i+1}] {book['metadata']['filename']}: {raw_size} -> {clean_size} chars (-{reduction:.1f}%)")

    print("\n✅ Library is clean")

# --- 3. RUN THE PROCESS ---
run_final_cleaning(all_books)
```

## Basic Deduplication

```python
import hashlib

def deduplicate_books(books_list):
    unique_books = []
    seen_hashes = set()

    print(f"Checking for duplicates in {len(books_list)} books...")
    
    for book in books_list:
        # Create a unique fingerprint based on the RAW content
        content_hash = hashlib.md5(book['content'].encode('utf-8')).hexdigest()

        if content_hash not in seen_hashes:
            unique_books.append(book)
            seen_hashes.add(content_hash)
        else:
            print(f"  ⚠ Skipping duplicate: {book['metadata']['filename']}")
    print(f"Deduplication complete. {len(unique_books)} unique books remaining.")
    return unique_books
```
